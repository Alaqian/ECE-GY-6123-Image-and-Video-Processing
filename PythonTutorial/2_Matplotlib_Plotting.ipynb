# Matplotlib Tutorial

```python
import numpy as np
import matplotlib.pyplot as plt
```

matplotlib.pyplot is a library for creating figures/plots and also display images

### First step to plotting
To create a plot, we need to use the function plt.plot()

plt.plot() takes two input, one for x-axis and one for y-axis. The two must be a list, or np array of same length.

For example, we want to plot y = sin(x) form 0 to 10

```python
## First create an array of x-axis values
x = np.arange(0,10,0.5)   # This means create an array with values from 0 to 10, with step size of 0.5
print(x)

y = np.sin(x)    # np.sin(x_axis) evaluate the sin() value of each element in x_axis array.
print(y)
```

```python
plt.figure()    # Creates a new figure
# Plot a function using the points in x and y, connect them with a straight line
plt.plot(x,y)    
```

```python
## Now use an x with smaller step size
x = np.arange(0,10,0.01) 
y = np.sin(x)    
plt.figure()
plt.plot(x,y)
## Note that this plot is smoother than the last one, because more points are used
```

```python
y2 = np.sqrt(x)
## plot multiple functions
plt.figure()
plt.plot(x,y, label = 'sin(x)')
plt.plot(x,y2, label = 'sqrt(x)')
plt.legend()    # Turn on legend
plt.grid()      # Turn on grid
plt.xlabel('Time')
plt.ylabel('Voltage')   ## Labels for x and y axis
```

### More options for plotting
- There are many options for customizing the plot, check out the matplotlib documentation online.

```python
x = np.arange(0,10)
y = np.random.randn(10)
y2 = np.random.randint(1,5,(10))
ax = plt.figure()
plt.plot(x,y, '--', color = 'b')   # Choose line style and color
plt.plot(x,y2,':', color = 'r')
```
