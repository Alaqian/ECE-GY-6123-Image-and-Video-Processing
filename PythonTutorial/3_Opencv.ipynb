## CV2 Tutorial

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### Reading image

```python
## Read image from .jpg file. Make sure the image is in the same directory as notebook
img = cv2.imread('Berries.jpg')    
print(type(img))  # img is a 3-dimensional numpy array
print(img.shape)  # First two dimensions are the image spatial size, last dimension is number of channels
```

```python
plt.figure()
plt.imshow(img)
# The color of the image being displayed here is not the same as original .jpg file
# This is because cv2 reads image in BGR, but plt.imshow() display image in RGB
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)   #converts color space from BGR to RGB
plt.figure()
plt.imshow(img_rgb)
```

### Different color channels of an image

```python
## Look at the intensity of each color channel
img_red = img_rgb[:,:,0]
img_green = img_rgb[:,:,1]
img_blue = img_rgb[:,:,2]
plt.figure(figsize = (15,15))
plt.subplot(1,3,1)
plt.imshow(img_red, cmap = 'gray')     # 'gray' is to show a single-channel image in grayscale
plt.title('R')
plt.subplot(1,3,2)
plt.imshow(img_green, cmap = 'gray')
plt.title('G')
plt.subplot(1,3,3)
plt.imshow(img_blue, cmap = 'gray')
plt.title('B')
```

```python
img_ycrcb = cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)    # Change to YCrCb 
img_gray = img_ycrcb[:,:,0]   # The first channel Y is luma component. Or the grayscale of a color image
plt.figure()
plt.imshow(img_gray, 'gray')
```

### Save image into .jpg file

```python
# Write the gray-scale image to a file
cv2.imwrite('Berries_grayscale.jpg', img_gray)
```

### Image filtering using cv2

```python
kernel = np.ones((5,5))/25   ## Define a uniform kernel as a numpy array
print(kernel)
img_blurred = cv2.filter2D(img_gray,-1,kernel)
# -1 is used to get the output same size as the input
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img_gray,'gray')
plt.subplot(1,2,2)
plt.imshow(img_blurred, 'gray')
```
